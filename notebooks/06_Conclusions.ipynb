## Opening Summary

This project revealed that mobile money fraud is not random but a structured two-stage crime: fraudsters first use TRANSFER transactions to move stolen funds from high-value victim accounts (draining 88% of balances on average), then complete the theft through CASH_OUT transactions that convert digital money into physical cash. The analysis exposed that while the ecosystem operates with clear "normal" patterns dominated by CASH_OUT for liquidity and PAYMENT for daily activity fraudsters actively exploit these same patterns, testing system limits with zero-value transactions and deliberately varying amounts to avoid detection. Most critically, the current fraud detection system fails catastrophically: despite a misleading 99.87% accuracy, it catches only 0.19% of actual fraud (16 out of 8,213 cases), allowing 99.8% of fraudulent value over KSh 11.97 billion to flow through undetected while maintaining a perfect false positive rate that prioritizes user experience over financial protection.

## Key Findings

- **Chapter 1:** The normal ecosystem is dominated by CASH_OUT and PAYMENT (69% of all transactions combined), but TRANSFER carries the highest average value (KSh 910,647) — fraudsters later exploit TRANSFER precisely because it is designed for high-value movement.

- **Chapter 2:** Fraud is 100% concentrated in TRANSFER and CASH_OUT — zero fraud occurs in PAYMENT, DEBIT, or CASH_IN — confirming fraud follows a deliberate two-stage sequence: move money first, then withdraw it.

- **Chapter 3:** Fraudulent transactions drain 88% of victim account balances on average, hit the exact system limit of KSh 10 million, and include zero-value probes — clear evidence fraudsters test and exploit system rules.

- **Chapter 4:** The top 8.99% of victims (those losing over KSh 5 million) account for roughly half of all fraud losses (KSh 6.06 billion) — fraud harm is concentrated, not evenly distributed.

- **Chapter 5:** The detection system catches only 16 out of 8,213 fraud cases (0.19% recall) — despite 99.87% accuracy, the system is functionally blind to almost all actual fraud.

## Business Recommendations

**Recommendation 1: Flag any TRANSFER or CASH_OUT that drains more than 80% of the sender's balance.**

Your data shows fraudsters take 88% of victim balances on average. Legitimate users do not drain almost their entire account in one transaction. Set a rule: if a TRANSFER or CASH_OUT reduces the sender's balance by more than 80%, hold it for manual review. This would have caught most of your fraud cases.

**Recommendation 2: Flag any transaction exactly at KSh 10,000,000.**

Fraudsters hit the system limit because they want to take as much as possible in one go. Regular users almost never send exactly the maximum allowed amount. Add one simple rule: any TRANSFER or CASH_OUT at KSh 10,000,000 gets flagged. No machine learning needed. This works today.

**Recommendation 3: Stop scanning PAYMENT, DEBIT, and CASH_IN for fraud.**

You have 6.36 million transactions. Zero fraud in those three types. You are wasting time and computing power checking transactions that never contain fraud. Remove detection rules from PAYMENT, DEBIT, and CASH_IN. Focus only on TRANSFER and CASH_OUT. That cuts your monitoring surface by more than half.

**Recommendation 4: Investigate large losses first, not many losses.**

Nine percent of victims (those losing over 5 million) account for half your total fraud loss in value. The other half of victims (those losing under 500,000) account for only 7% of total loss. When a case comes in, look at the amount. Big loss gets immediate freeze and an investigator. Small loss goes to a batch report at end of day. You cannot investigate everything equally. Prioritize by money.

**Recommendation 5: Watch for mule accounts.**

Recipient accounts in fraud cases start with low balances (around KSh 544,000), receive a large transfer, then send the money out quickly. Build a simple score: low starting balance? Money leaves within 2 hours? Outgoing transactions split into small pieces? Any account that looks like this gets flagged before it receives another transfer. This stops fraud before the second transaction happens.

## Fraud Detection Strategy

**Step 1: Change what you measure.**

Stop looking at accuracy. Every morning check: how many fraud cases did we catch? How many legitimate customers did we block? How much money did we lose?

**Step 2: Run three simple rules on TRANSFER and CASH_OUT only.**

- Flag any transaction that drains more than 80% of the sender's balance
- Flag any transaction exactly at KSh 10,000,000
- Flag any recipient account with low balance (under KSh 600,000) receiving a large transfer (over KSh 1,000,000)

**Step 3: Start with daily manual review.**

Do not block transactions automatically at first. Run the rules every morning. Review the results manually. Track how many flags turn out to be fraud. Learn from what you find.

**Step 4: Stop monitoring PAYMENT, DEBIT, and CASH_IN.**

Fraud never happens in these transaction types. Remove them from your detection queries. Focus only on where fraud actually occurs.

**Step 5: Prioritize by loss amount.**

When you find a potential fraud case, look at the amount. Big loss gets investigated first. Small loss goes to the end of the queue.

## Project Limitations and Next Steps

**Limitations**

PaySim is synthetic data, not real M-Pesa transactions. Three things it cannot tell you. First, real fraudster behavior is messier and adapts faster than a simulation. PaySim fraud patterns are clean and predictable. Real fraud will not be. Second, the data has no customer information no age, location, account tenure, or device type. In a real fraud job, you would need to know if victims share something in common (elderly users? certain regions? specific phone models?) to target prevention. Third, the time granularity is one hour. Real detection requires minute level or second level data to catch mule accounts that move money within minutes, not hours. PaySim shows you patterns. Real data shows you speed.
